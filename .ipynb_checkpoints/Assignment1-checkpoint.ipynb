# Question One

## Part 1.1

```python
def construct_period(increment, duration_ones):
    period = []
    for i in range(increment):
        if i < duration_ones:
            period.append(1)
        else:
            period.append(0)
    return period
```

```python
def construct_TC(onset_arrival, increment, duration_ones):
    TC = []
    for i in range(onset_arrival):
        TC.append(0)
    while(len(TC) < 240):
        for val in construct_period(increment, duration_ones):
            TC.append(val)
    
    # Only take the first 240
    return TC[:240]
```

#### Construct TCs

```python
TC_list = []
AV = [0, 20, 0, 0, 0, 0]
IV = [30, 45, 60, 40, 40, 40]
Ones = [15, 20, 25, 15, 20, 25]
for i in range(6):
    TC_list.append(construct_TC(AV[i], IV[i], Ones[i]))
```

#### Transpose (to 240 x 6) and Standardise


```python
TC = np.transpose(np.array(TC_list))
TC = StandardScaler().fit_transform(TC)
```

#### Plot TCs

```python
plt.subplots(2, 3, figsize=(15,10))
for i in range(6):
    tc = TC[:, i]
    plt.subplot(2, 3, i+1)
    plt.title(f"TC {i+1}")
    plt.plot(tc)    
plt.savefig("plots/1.1 TC subplots.png")
plt.show()
```

## Part 1.3

#### Construct TmpSM (6 x (21 x 21))

```python
def constructTmpSM(v1, v2, h1, h2):
    v1 -= 1
    v2 -= 1
    h1 -= 1
    h2 -= 1
    tmp = []
    for i in range(21):
        row = []
        for j in range(21):
            num_to_add = 0
            if i >= v1 and i <= v2:
                if j >= h1 and j <= h2:
                    num_to_add = 1
            row.append(num_to_add)
#             tmp.append(num_to_add)
        tmp.append(row)
    
    return tmp
```

```python
slices = [
    (2, 6, 2, 6),
    (2, 6, 15, 19),
    (8, 13, 2, 6),
    (8, 13, 15, 19),
    (15, 19, 2, 6),
    (15, 19, 15, 19)
]
```

```python
array = []
for slice_ in slices:
    array.append(constructTmpSM(slice_[0], slice_[1], slice_[2], slice_[3]))
tmpSM = np.array(array)
```

#### Plot SMs in six subplots

## Part 2.3

```python
average_mse_list = []
rhos = []
# For each value of rho, create 10 random realisations of X and obtain Alr and Dlr
for rho in np.arange(0.0, 1.05, 0.05):
    
    mse_list = []
    for i in range(10):
        # reset the seed, generate noise
#         np.random.seed()
        temp_noise_mu = 0
        temp_noise_var = 0.25

        spat_noise_mu = 0
        spat_noise_var = 0.015

        temp_noise = np.random.normal(temp_noise_mu, math.sqrt(temp_noise_var), 240*6)
        spatial_noise = np.random.normal(spat_noise_mu, math.sqrt(spat_noise_var), 6*441)

        Gamma_t = temp_noise.reshape(240, 6, order="F")
        Gamma_s = spatial_noise.reshape(6, 441, order = "F")
        
        # Recalculate X, and standardise
        tmpX = np.dot((TC + Gamma_t), (SM + Gamma_s))     
        tmpX = StandardScaler().fit_transform(tmpX)
        
        
        N = 240
        nsrcs = 6
        x1 = 21
        x2 = 21

        ###################################################################
        ## Code for LR in python. Supplied by the University Of Melbourne in R and then translated by myself to Python
        step = 1/(np.linalg.norm(np.dot(TC, np.transpose(TC))) * 1.1)
        thr = rho * N * step
        Ao = np.zeros((nsrcs, 1))
        A = np.zeros((nsrcs, 1))
        Alr = np.zeros((nsrcs, x1*x2))
        for k in range(1, x1*x2+1):

            A = Ao + step*(np.dot(np.transpose(TC), tmpX[:, k-1].reshape(N,1) - np.dot(TC, Ao)))
            A = (1/(1+thr)) * np.sign(A) * np.maximum(np.zeros((nsrcs, 1)), (abs(A) - thr))
            for i in range(1, 11):
                Ao = A
                A = Ao + step*(np.dot(np.transpose(TC), tmpX[:, k-1].reshape(N,1) - np.dot(TC, Ao)))
                A = (1/(1+thr)) * np.sign(A) * np.maximum(np.zeros((nsrcs, 1)), (abs(A) - thr))     
            Alr[:,k-1] = A.reshape(6)

        ###################################################################

        Dlr = np.dot(tmpX, np.transpose(Alr))

        ###################################################################
        ## Code for MSE in python. Supplied by the University Of Melbourne in R and then translated by myself to Python
        mse = sum(sum(np.dot(np.transpose(tmpX - np.dot(Dlr, Alr)), tmpX - np.dot(Dlr, Alr)))) / (N * x1*x2)
        ###################################################################
        mse_list.append(mse)
    average_mse_list.append(np.mean(mse_list))
    rhos.append(rho)
        
```

```python
plt.plot(rhos,average_mse_list)
plt.ylabel('Average MSE')
plt.xlabel('Rho')

plt.savefig("plots/2.3 rho against MSE.png")
plt.show()
```

## Part 2.4

#### Using $\rho = 0.6$ obtained above, we calculate Dlr and Alr using the same X that we used with RR and SLR

```python
rho = 0.60
###################################################################
## Code for LR in python. Supplied by the University Of Melbourne in R and then translated by myself to Python
step = 1/(np.linalg.norm(np.dot(TC, np.transpose(TC))) * 1.1)
thr = rho * N * step
Ao = np.zeros((nsrcs, 1))
A = np.zeros((nsrcs, 1))
Alr = np.zeros((nsrcs, x1*x2))
for k in range(1, x1*x2+1):

    A = Ao + step*(np.dot(np.transpose(TC), X[:, k-1].reshape(N,1) - np.dot(TC, Ao)))
    A = (1/(1+thr)) * np.sign(A) * np.maximum(np.zeros((nsrcs, 1)), (abs(A) - thr))
    for i in range(1, 11):
        Ao = A
        A = Ao + step*(np.dot(np.transpose(TC), X[:, k-1].reshape(N,1) - np.dot(TC, Ao)))
        A = (1/(1+thr)) * np.sign(A) * np.maximum(np.zeros((nsrcs, 1)), (abs(A) - thr))     
    Alr[:,k-1] = A.reshape(6)

###################################################################

Dlr = np.dot(X, np.transpose(Alr))
```

#### Apply Lasso with $\rho = 0.001$ using Z

```python
rho = 0.001
###################################################################
## Code for LR in python. Supplied by the University Of Melbourne in R and then translated by myself to Python
step = 1/(np.linalg.norm(np.dot(Z, np.transpose(Z))) * 1.1)
thr = rho * N * step
Ao = np.zeros((nsrcs, 1))
A = np.zeros((nsrcs, 1))
Alr = np.zeros((nsrcs, x1*x2))
for k in range(1, x1*x2+1):

    A = Ao + step*(np.dot(np.transpose(Z), X[:, k-1].reshape(N,1) - np.dot(Z, Ao)))
    A = (1/(1+thr)) * np.sign(A) * np.maximum(np.zeros((nsrcs, 1)), (abs(A) - thr))
    for i in range(1, 11):
        Ao = A
        A = Ao + step*(np.dot(np.transpose(Z), X[:, k-1].reshape(N,1) - np.dot(Z, Ao)))
        A = (1/(1+thr)) * np.sign(A) * np.maximum(np.zeros((nsrcs, 1)), (abs(A) - thr))     
    Alr[:,k-1] = A.reshape(6)

###################################################################

Apcr = Alr
Dpcr = np.dot(X, np.transpose(Alr))
```
